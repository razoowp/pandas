## set_index() , inplace=True

```python
## Setting day as index. inplace=True will replace the original dataframe and doesn't create new one.
import pandas as pd
df = pd.read_csv("weather_data_missingdata.csv", parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

## fillna()

```python
new_df = df.fillna(0)
new_df
```

```python
## we don't want to fill just 0, we can use dictionary to fill the customize data
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'No Event'
})
new_df
```

```python
# we can use: ffill or bfill to fill NaN cell. ffill from gets data from top and bfill from bottom.
# This is vertically, if we want horizontally we use another attribute: axis="columns"

## see pandas fillna documentation

df_new = df.fillna(method="ffill")
df_new
```

### limit attribute will limit how much cell to be copied with ffill and bfill eg. limit=1

## interpolate()

```python
## interpolate() will come up with better guess. check pandas doc for other types of interpolation.thi is linear.
# method="time" gives more appropriate guess considering date gaps.
new_df = df.interpolate(method="time")
new_df
```

## dropna()

```python
#Drops all the rows having NaN values
df_new = df.dropna()
df_new
```

```python
# how="all" removes the row without any values in all fields. 
# thresh=1 will keep the row having at least one non na values.
df_new = df.dropna(how="all")
df_new
```
