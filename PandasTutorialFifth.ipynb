## Handling Missing data - replace function

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data_replace.csv")
df
```

```python
#df_new = df.replace(-99999, np.NaN)
df_new = df.replace([-99999, -8888], np.NaN)
df_new
```

```python
## to replace with specific columns we use DICTIONARY
df_new = df.replace({
    'temperature' : -99999,
    'windspeed' : [-99999, -8888],
    'event' : '0'
}, np.NaN)
df_new
```

```python
df_new = df.replace({
    -99999 : np.NaN,
    '0' : "no event"
})
df_new
```

```python
## Now we have changed the csv file for some more interesting topic. temperature has F and windspeed has mph
df = pd.read_csv("weather_data_replace.csv")
df
```

```python
## This will replace all text associated with value within temperature and winspeed column USING REGULAR EXPRESSION
new_df = df.replace({
    'temperature' : '[a-zA-z]',
    'windspeed' : '[a-zA-Z]'
}, '',regex=True)
new_df
```
